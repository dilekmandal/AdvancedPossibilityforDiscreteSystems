**PROBLEM:** Bir okulun öğrencilerinden %45’i fizik, %35’i matematik derslerinden ve %25’i hem fizik hem de matematik derslerinden başarısızdır. Rasgele seçilen bir öğrencinin,

a) Fizikten başarısız olduğu biliniyorsa, matematikten de başarısız olma olasılığını bulunuz

b)Matematikten başarısız olduğu biliniyor ise, fizikten de başarısız olma olasılığını bulunuz

**ÇÖZÜM:**

image.png

```python
print("Fizik dersinden başarısız olanların oranını girin: ")
pF = float(input("P(F): "))
print("Matematik dersinden başarısız olanların oranını girin:")
pM = float(input("P(M): "))
print("Her ikisinden de başarısız olanların oranını girin:")
pFnM = float(input("P(F n M): "))

pFM = pFnM / pM
pMF = pFnM / pF

print("Fizik dersinden başarısız olanların Matematik dersinden de başarısız olma olasılığı => p(F|M) :", pFM)
print("Matematik dersinden başarısız olanların Fizik dersinden de başarısız olma olasılığı => p(M|F) :", pMF)
```

**BAYES TEOREMİ**

P(A|B) = B olayı gerçekleştiğinde A olayının gerçekleşme olasılığı

P(A) = A olayının gerçekleşme olasılığı

P(B|A) = A olayı gerçekleştiğinde B olayının gerçekleşme olasılığı

P(B) = B olayının gerçekleşme olasılığı

**SORU:**
Diyabet testi alan insanların %20'sinin gerçekten diyabet hastası olduğu tespit edilmiştir. Belli bir markanın test sonuçlarına göre;
Testi yapan kişi gerçekten diyabet hastasıysa, %90 ihtimalle test pozitif sonuç vermektedir.Testi yapan kişi diyabet hastası değilse, %80 ihtimalle test negatif sonuç vermektedir. Tüm bu veriler dikkate alınarak, bir kişi testi almışsa ve sonuç pozitifse bu kişinin gerçekten diyabet hastası olma olasılığı nedir?

**ÇÖZÜM:**

B1 : Diyabet hastası olanlar = 0,2

B2 : Diyabet hastası olmayanlar = 0,8

A  : Test sonucu pozitif olanlar

A|B1 : Test sonucu pozitif olup diyabet hastası olanlar = 0,9

A|B2 : Test sonucu pozitif olup diyabet hastası olmayanlar  = 0,8

B1|A : Diyabet hastası olup testi pozitif çıkanlar

P(B1|A) = P(A|B1)*P(B1) / ((P(A|B1)*P(B1)) + (P(A|B2)*P(B2)))


P(B1|A )= (0.9*0.2) / ((0.9*0.2)+(0.2*0.8)) = 9/17



```python
def BayesTheorem(ab1, ab2, b1, b2):
  return b1*ab1 / (b1*ab1 + b2*ab2)

pB1 = 0.20
pB2 = 0.80
pAB1 = 0.90
pAB2 = 0.20

print("Test sonucu pozitif çıkan bir kişinin gerçekten diyabet hastası olma ihtimali : ", BayesTheorem(pAB1, pAB2, pB1, pB2))
```
