```python
def kesikli_olasilik_fonksiyonu(x):
  if x>= 1 and x<=10:
    return x*(1/100)      
  elif x>= 11 and x<=20:
    return (20-x)*(1/100)

print("a) Balıkçının tam 8 balık tutma olasılığını hesaplayın.")
print("Balıkçının tam 8 balık tutma ihtimali = P(8) :",kesikli_olasilik_fonksiyonu(8))
print("\n")


print("b) Balıkçının 8'den az balık tutma olasılığını hesaplayınız")

sekizden_az_balik = 0
i = 0
while i < 7:
  i += 1
  
  print("Balıkçının",i," balık tutma ihtimali = P(",i,"): ",kesikli_olasilik_fonksiyonu(i))
  sekizden_az_balik += kesikli_olasilik_fonksiyonu(i)
print("\n")
print("Balıkçının 8'den az balık  tutma ihtimali : ",sekizden_az_balik)



print("\n")
print("c) Balıkçının 8'den çok balık tutma olasılığını hesaplayınız")
sekizden_cok_balik = 0
i = 8
while i < 20:
  i += 1
  
  print("Balıkçının",i," balık tutma ihtimali = P(",i,"): ",kesikli_olasilik_fonksiyonu(i))
  sekizden_cok_balik += kesikli_olasilik_fonksiyonu(i)
print("\n")
print("Balıkçının 8'den cok balık  tutma ihtimali : ",sekizden_cok_balik)


print("\n")
print("d) Balıkçının 6 ile 18 arasında balık tutma olasılığını hesaplayınız")

total = 0
i = 5
while i < 18:
  i += 1
  
  print("Balıkçının",i," balık tutma ihtimali = P(",i,"): ",kesikli_olasilik_fonksiyonu(i))
  total += kesikli_olasilik_fonksiyonu(i)
print("\n")
print("Balıkçının 6 ile 18 arasında balık  tutma ihtimali : ",total)
```
